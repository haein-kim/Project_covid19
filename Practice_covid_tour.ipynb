```python
covid = pd.read_excel('practice_covid.xlsx')
covid.head()
```

```python
covid.columns = ["Date", "확진자 수"]
covid.head()
```

## tour 공휴일 채우기

```python
# 범위를 정하기에 앞서서 첫날 과 마지막날 정의하기
tour_first_day = tour.iloc[0, 0]
tour_last_day = tour.iloc[252, 0]
tour_first_day, tour_last_day
```

```python
# 범위 함수로 빠진 날짜까지 불러오기
tour_days = pd.date_range(tour_first_day, tour_last_day)
tour_days
```

```python
# 데이터프레임으로 넣어주기
df_tour_days = pd.DataFrame({"날짜": tour_days})
df_tour_days.head()
```

```python
tour.head()
```

```python
# 합치기
tour_all_day = df_tour_days.merge(tour, left_on="날짜", right_on=tour["Date"], how="left")
tour_all_day[:5]
```

```python
tour_all_day[["노랑풍선","모두투어","하나투어","참좋은여행","용평리조트","아난티","롯데관광개발","GKL","강원랜드","호텔신라"]] = tour_all_day[["노랑풍선","모두투어","하나투어","참좋은여행","용평리조트","아난티","롯데관광개발","GKL","강원랜드","호텔신라"]].fillna(method="bfill")
tour_all_day[:5]
```

```python
tour = tour_all_day[["날짜", "노랑풍선","모두투어","하나투어","참좋은여행","용평리조트","아난티","롯데관광개발","GKL","강원랜드","호텔신라"]]
tour.head()
```

## tour + covid Merge

```python
tour_covid = tour.merge(covid, left_on="날짜", right_on=covid["Date"], how="left")
tour_covid.head()
```

```python
tour_covid["확진자 수"] = tour_covid["확진자 수"].fillna(0)
tour_covid[:20]
```

* 저장하기

```python
tour_covid.to_excel("tour_covid.xlsx", index=False, encoding="utf8")
tour_covid.to_pickle("tour_covid.pkl")
tour_covid.to_html()
tour_covid.to_html("tour_covid.html", index=False)
```
